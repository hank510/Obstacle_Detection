```python
# For high pass filtered photos
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D

import numpy as np
import os
from keras_preprocessing import image
# from tensorflow.keras.preprocessing.image import load_img
# from tensorflow.keras.utils import load_img
True_count_No = 0
False_count_No = 0

True_count_O = 0
False_count_O = 0

model=Sequential()
model.add(Conv2D(16, kernel_size=(5, 5),activation='relu',input_shape=(150,150,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, kernel_size=(5, 5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

path = 'C:\\Users\\Vrushank Anand\\Documents\\College Related\\Semester 7\\Project_Ykm_Sir\\Codes_for_project\\object_detection_normalized_high_pass_filter_model.h5'
model.load_weights(path)

folder_O = 'C:\\Users\\Vrushank Anand\\Documents\\College Related\\Semester 7\\Project_Ykm_Sir\\Normalized_High_Pass_Filtered_Photos\\val\\Object\\'
for filename in os.listdir(folder_O):
    test_image =image.load_img(os.path.join(folder_O,filename),target_size =(150,150))
    test_image =image.img_to_array(test_image)
    test_image =np.expand_dims(test_image, axis =0)
    
    result = model.predict(test_image)
    if result[0][0] >= 0.5:
        prediction = 'Object'
        True_count_O = True_count_O + 1
    else:
        prediction = 'No object'
        False_count_O = False_count_O + 1
    print(prediction, filename)


folder_No = 'C:\\Users\\Vrushank Anand\\Documents\\College Related\\Semester 7\\Project_Ykm_Sir\\Normalized_High_Pass_Filtered_Photos\\val\\No object\\'
for filename in os.listdir(folder_No):
    test_image =image.load_img(os.path.join(folder_No,filename),target_size =(150,150))
    test_image =image.img_to_array(test_image)
    test_image =np.expand_dims(test_image, axis =0)
    
    result = model.predict(test_image)
    if result[0][0] >= 0.5:
        prediction = 'Object'
        False_count_No = False_count_No + 1
    else:
        prediction = 'No object'
        True_count_No = True_count_No + 1
    print(prediction, filename)

print('True No Object counter', True_count_No)
print('False No Object counter', False_count_No)

print('True Object counter', True_count_O)
print('False Object counter', False_count_O)

efficiency_No_obj = (True_count_No/(True_count_No+False_count_No))*100
efficiency_Obj = (True_count_O/(True_count_O+False_count_O))*100
efficiency_Net = ((True_count_No+True_count_O)/(True_count_No+False_count_No+True_count_O+False_count_O))*100

print('True No Object efficiency', efficiency_No_obj)
print('True Object efficiency', efficiency_Obj)
print('Total Efficiency', efficiency_Net)
```
